```python
import PyPDF2, urllib.request, nltk, textract
from textract import process
from io import BytesIO
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import sent_tokenize
from nltk.tokenize import BlanklineTokenizer
import pandas as pd
import PyPDF2
import datafunc
import csv
import re
from collections import Counter
from  itertools import chain

nltk.download('stopwords')
```

```python
reader = PyPDF2.PdfFileReader('17141.pdf')
num_of_pages = reader.numPages
```

```python
def getPDFContent(path,num_of_pages):
    page_list =[]
    content = ""
    num_pages = num_of_pages
    p = open(path, "rb")
    pdf = PyPDF2.PdfFileReader(p)
    for i in range(0, num_pages):
        page_list.append(pdf.getPage(i).extractText())
        
    return page_list 


content = getPDFContent('17141.pdf',num_of_pages)
```

```python
content[1]
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(str(content))
```

```python
sentences_list = []
for sentence in sentences:
    sentences_list.append(sentence.replace('\\n', ''))

df = pd.DataFrame(sentences_list)
df = df.rename(columns={0:"Sentence"})
df
```

```python
sentences_list[3]
```

```python
keywords = pd.read_excel(io = 'keywords.xlsx', sheet_name='Sheet1')
Keywords_list = keywords["Keywords"].to_list() 
for Keyword in Keywords_list:
    df[Keyword] = df["Sentence"].str.findall(Keyword, flags=re.IGNORECASE)
```

```python
list_1 = []
for sentence in df.Sentence:
    list_2 = []
    
    for keyword in Keywords_list:
        
        list_2.append(sentence.count(keyword))
                      
    
    list_1.append(list_2)
list_1
```
